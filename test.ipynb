```python
from ultralytics import YOLO
import cv2

# Load the model
model = YOLO('offline_exam_monitoring_model.pt')

# Load the image
image_path = 'image.jpeg'
image = cv2.imread(image_path)

# Perform inference
results = model(image)

# Print the results
for result in results[0].boxes:  # Access the first image's results
    label = int(result.cls)
    confidence = result.conf
    bbox = result.xyxy
    print(f"Label: {model.names[label]}, Confidence: {confidence}, Bounding Box: {bbox}")

# Optionally, visualize the results
annotated_image = results[0].plot()
cv2.imshow('Annotated Image', annotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

# Load the model
model = YOLO('offline_exam_monitoring_model.pt')

# Initialize camera index
camera_index = 0

def switch_camera():
    global camera_index
    camera_index = 1 - camera_index  # Toggle between 0 and 1

# Open the camera
cap = cv2.VideoCapture(camera_index)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Perform inference
    results = model(frame)

    # Annotate the frame
    annotated_frame = results[0].plot()

    # Display the frame
    cv2.imshow('Live Detection', annotated_frame)

    # Check for key presses
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):  # Press 'q' to quit
        break
    elif key == ord('s'):  # Press 's' to switch camera
        cap.release()
        switch_camera()
        cap = cv2.VideoCapture(camera_index)

# Release the camera and close windows
cap.release()
cv2.destroyAllWindows()
```
